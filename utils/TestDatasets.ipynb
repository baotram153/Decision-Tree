```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("outsiders17711/2d-datasets")

print("Path to dataset files:", path)
```

```python
import os
from matplotlib import pyplot as plt
import pandas as pd

path_based = os.path.join(path, "Pathbased.txt")
aggregation = os.path.join(path, "Aggregation.txt")
compound = os.path.join(path, "Compound.txt")
jain = os.path.join(path, "Jain.txt")
flame = os.path.join(path, "flame.txt")
spiral = os.path.join(path, "spiral.txt")


path_based_df = pd.read_csv(path_based, skiprows=7, delimiter='\t', header=None)
aggregation_df = pd.read_csv(aggregation, skiprows=7, delimiter='\t', header=None)
compound_df = pd.read_csv(compound, skiprows=7, delimiter='\t', header=None)
jain_df = pd.read_csv(jain, skiprows=7, delimiter='\t', header=None)
flame_df = pd.read_csv(flame, skiprows=7, delimiter='\t', header=None)
spiral_df = pd.read_csv(spiral, skiprows=7, delimiter='\t', header=None)
```

```python
def preprocess_scatter(df):
    X = df[[0, 1]]
    y = df[[2]]
    return X, y
```

```python
X, y = preprocess_scatter(path_based_df)
plt.scatter(x=X[[0]], y=X[[1]], c=y)
```

```python
X, y = preprocess_scatter(aggregation_df)
plt.scatter(x=X[[0]], y=X[[1]], c=y)
```
